```python
import pandas as pd
import numpy as np
from sklearn.preprocessing import LabelEncoder, MinMaxScaler
from sklearn.ensemble import RandomForestClassifier
from sklearn.model_selection import train_test_split
```

```python
# Load the training dataset
train_data = pd.read_csv('/home/arnab/Desktop/Doceree-HCP_Train.csv',encoding = "latin-1")
train_data.head()
```

```python
# Preprocess the training data
train_data = train_data.drop('CHANNELTYPE', axis=1)  # Drop the 'CHANNELTYPE' column
train_data = train_data.drop('ID', axis=1)  # Drop the 'id' column
train_data = train_data.fillna('Unknown')  # Fill missing values with 'Unknown'
taxonomy_data=train_data.drop('IS_HCP',axis=1)
hcp_data=train_data.drop('TAXONOMY',axis=1)
```

```python
hcp_data.head()
```

```python
taxonomy_data['TAXONOMY'].value_counts()
```

```python
#Encoding text to numeric data
le = LabelEncoder()
for col in hcp_data.columns:
    hcp_data[col] = le.fit_transform(hcp_data[col].astype(str))  
hcp_data.head()
```

```python
y_taxonomy=np.array(taxonomy_data['TAXONOMY'])
taxonomy_data=taxonomy_data.drop('TAXONOMY',axis=1)  
taxonomy_data.head()
```

```python
le1 = LabelEncoder()
for col in taxonomy_data.columns:
    taxonomy_data[col] = le1.fit_transform(taxonomy_data[col].astype(str))  
taxonomy_data.head()
```

```python
le2 = LabelEncoder()
y_taxonomy= le2.fit_transform(y_taxonomy.astype(str))  
y_taxonomy
```

```python
y_hcp=hcp_data['IS_HCP']
y_hcp
```

```python
print(y_taxonomy)
type(y_taxonomy)
```

```python
print(y_hcp)
type(y_hcp)
```

```python
# Split the data into train and validation sets
X_hcp=hcp_data.drop('IS_HCP',axis=1)

X_train_hcp, X_val_hcp, y_train_hcp, y_val_hcp= train_test_split(
    X_hcp, y_hcp, test_size=0.2, random_state=42
)

X_train_taxonomy, X_val_taxonomy, y_train_taxonomy, y_val_taxonomy = train_test_split(
    taxonomy_data, y_taxonomy, test_size=0.2, random_state=42
)
```

```python
# Build the IS_HCP model (Random Forest Classifier)
model_hcp = RandomForestClassifier(n_estimators=100, random_state=42)
model_hcp.fit(X_train_hcp, y_train_hcp)
```

```python
#Model Scores
print('Train Error:')
print(1-model_hcp.score(X_train_hcp, y_train_hcp))
print('Validation Error:')
print(1-model_hcp.score(X_val_hcp, y_val_hcp))
```

```python
# Build the TAXONOMY model (Random Forest Classifier)
model_taxonomy = RandomForestClassifier(n_estimators=100, random_state=42)
model_taxonomy.fit(X_train_taxonomy, y_train_taxonomy)
```

```python
#Model Scores
print('Train Error:')
print(1-model_taxonomy.score(X_val_taxonomy,y_val_taxonomy))
print('Validation Error:')
print(1-model_taxonomy.score(X_val_taxonomy,y_val_taxonomy))
```

```python
# Load the testing dataset
test_data = pd.read_csv('/home/arnab/Desktop/Doceree-HCP_Test.csv',encoding="latin-1")
test_data
```

```python
# Preprocess the testing data
test_data = test_data.drop('CHANNELTYPE', axis=1)  # Drop the 'CHANNELTYPE' column
id_col=test_data['ID']#Saving the original id
test_data = test_data.drop('ID', axis=1)  # Drop the 'ID' column
test_data = test_data.fillna('Unknown')  # Fill missing values with 'Unknown'
test_hcp_data=test_data
test_taxo_data=test_data
test_data.head()
```

```python
for col in test_hcp_data.columns:
    test_hcp_data[col] = le.fit_transform(test_hcp_data[col].astype(str))    

for col in test_taxo_data.columns:
    test_taxo_data[col] = le1.fit_transform(test_taxo_data[col].astype(str))        
```

```python
# Step 9: Make predictions on the testing data
y_pred_hcp = model_hcp.predict(test_hcp_data)
print(y_pred_hcp)
y_pred_taxonomy = model_taxonomy.predict(test_taxo_data)
print(y_pred_taxonomy)
len(y_pred_taxonomy)
```

```python
predictions_taxonomy = le2.inverse_transform(y_pred_taxonomy)
print(len(predictions_taxonomy))
np.unique(predictions_taxonomy,return_counts=True)
```

```python
#Examining cases where taxonomy and hcp clashes
for i in range(len(predictions_taxonomy)):
    if(y_pred_hcp[i]==0):
        if(predictions_taxonomy[i]=='Unknown'):
            continue
        else:
            predictions_taxonomy[i]='Unknown'
for i in range(len(predictions_taxonomy)):
    if(predictions_taxonomy[i]=='Unknown'):
        predictions_taxonomy[i]=''           
```
